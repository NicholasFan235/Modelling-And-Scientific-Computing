# Solve Initial Value Problems

```python
import scipy.integrate
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

$\frac{dy}{dx} = x^2$, $y(0)=1$

```python
solution = np.empty((40, 2), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(0, 10, solution.shape[0])
analytic[:, 0] = np.linspace(0, 10, analytic.shape[0])

analytic[:, 1] = analytic[:, 0]**3 / 3 + 1
solution[:, [1]] = scipy.integrate.odeint(
    lambda y, x: x*x, 1, solution[:, 0])

plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 1], 'rx')
plt.show()
```

$\frac{dx}{dt}=\frac{t^2}{x}$, $x(0)=1$

```python
solution = np.empty((40, 2), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(0, 100, solution.shape[0])
analytic[:, 0] = np.linspace(0, 100, analytic.shape[0])

analytic[:, 1] = np.sqrt(2/3 * analytic[:, 0]**3 + 1)
solution[:, [1]] = scipy.integrate.odeint(
    lambda x, t: t*t/x, 1, solution[:, 0])

plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 1], 'rx')
plt.show()
```

$\frac{dy}{dx} + \frac{y}{x} = 1, y(1) = 0$

```python
solution = np.empty((40, 2), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(1, 10, solution.shape[0])
analytic[:, 0] = np.linspace(1, 10, analytic.shape[0])

analytic[:, 1] = analytic[:, 0]/2 - 1/analytic[:, 0]/2
solution[:, [1]] = scipy.integrate.odeint(
    lambda y, x: 1-y/x, 0, solution[:, 0])

plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 1], 'rx')
plt.show()
```

$\frac{dy_1}{dx}=-y_2, \frac{dy_2}{dx}=y_1, y_1(0)=1, y_2(0)=0$

```python
solution = np.empty((40, 3), dtype=float)
analytic = np.empty((1000, 3), dtype=float)
solution[:, 0] = np.linspace(0, 5, solution.shape[0])
analytic[:, 0] = np.linspace(0, 5, analytic.shape[0])

analytic[:, 1] = np.cos(analytic[:, 0])
analytic[:, 2] = np.sin(analytic[:, 0])
solution[:, [1, 2]] = scipy.integrate.odeint(
    lambda y, x: (-y[1], y[0]), (1, 0), solution[:, 0])

plt.plot(analytic[:, 1], analytic[:, 2], 'b')
plt.plot(solution[:, 1], solution[:, 2], 'rx')
plt.show()
```

$\frac{d^2y}{dx^2}+3\frac{dy}{dx}-4y=0, y(0)=1, y'(0)=0$

```python
solution = np.empty((40, 3), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(0, 1, solution.shape[0])
analytic[:, 0] = np.linspace(0, 1, analytic.shape[0])

analytic[:, 1] = 1/5 * np.exp(-4*analytic[:, 0]) + 4/5 * np.exp(analytic[:, 0])
solution[:, [1, 2]] = scipy.integrate.odeint(
    lambda y, x: (4*y[1]-3*y[0], y[0]), (0, 1), solution[:, 0])

plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 2], 'rx')
plt.show()
```
