# Solve Boundary Value Problems

```python
import numpy as np
import scipy.integrate
import matplotlib.pylab as plt
```

$\frac{d^2y}{dx^2}+3\frac{dy}{dx}-4y=0$

```python
def func(x, y):
    return np.vstack((y[1], -3*y[1]+4*y[0]))
```

$y(0)=1, y(1)=1$

```python
solution = np.empty((40, 3), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(0, 1, solution.shape[0])
analytic[:, 0] = np.linspace(0, 1, analytic.shape[0])
A = (np.exp(5)-np.exp(4))/(np.exp(5)-1)
B = (np.exp(4)-1)/(np.exp(5)-1)
analytic[:, 1] = A * np.exp(-4*analytic[:, 0]) + B * np.exp(analytic[:, 0])

sol = scipy.integrate.solve_bvp(
    func, lambda ya, yb: (ya[0]-1, yb[0]-1),
    solution[:, 0], np.ones((2, solution.shape[0])))
solution[:, [1,2]] = np.vstack((sol.y[0], sol.y[1])).T
```

```python
plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 1], 'rx')
plt.show()
```

$y'(0)=0, y(0)=1$

```python
solution = np.empty((40, 3), dtype=float)
analytic = np.empty((1000, 2), dtype=float)
solution[:, 0] = np.linspace(0, 1, solution.shape[0])
analytic[:, 0] = np.linspace(0, 1, analytic.shape[0])
A = np.exp(4)/(1+4*np.exp(5))
B = 4*np.exp(4)/(1+4*np.exp(5))
analytic[:, 1] = A * np.exp(-4*analytic[:, 0]) + B * np.exp(analytic[:, 0])

sol = scipy.integrate.solve_bvp(
    func, lambda ya, yb: (ya[1], yb[0]-1),
    solution[:, 0], np.ones((2, solution.shape[0])))
solution[:, [1,2]] = np.vstack((sol.y[0], sol.y[1])).T
```

```python
plt.plot(analytic[:, 0], analytic[:, 1], 'b')
plt.plot(solution[:, 0], solution[:, 1], 'rx')
plt.show()
```
