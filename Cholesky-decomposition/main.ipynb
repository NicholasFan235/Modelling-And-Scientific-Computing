# Test Multivariate Normal Distribution Class

```python
import numpy as np
from multivariate_normal import MultivariateNormal
from plotnine import *
import pandas as pd
```

```python
def custom_covariance(n, sigma1, sigma2):
    xs = np.arange(0, n)
    cov = sigma1**2 * np.exp(-xs[:, np.newaxis] - xs[:, np.newaxis].T)**2 / sigma2**2
    cov += 1e-5 * np.eye(n)
    return cov
```

```python
cov = custom_covariance(100, 1, 10)
distrib = MultivariateNormal(cov)
```
